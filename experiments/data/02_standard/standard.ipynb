# Combine Data Sources

Combine processed carbon intensity, building, and weather data to create a holistic picture, then seperate out into timestepped files. Clean up the data by removing shoddy rows (rows with no data in one of the categories).

```python
import os
import pandas as pd

directory = f"{os.getcwd()}/.."

building_real_data = f"{directory}/01_preprocess/building_real__2023.parquet"

building_real_df = pd.read_parquet(building_real_data)
```

```python
building_real_df.index = pd.DatetimeIndex(building_real_df.index, tz="Europe/Dublin")
building_real_df.head()
```

```python
combined_df = building_real_df
combined_df.head()
```

```python
"""
Any rows with no data for one of the data sources should be removed.
"""
combined_df_clean = combined_df[0:8760].copy()
combined_df_clean.head()
```

```python
import os
cwd = os.getcwd()
combined_real_df = combined_df_clean[building_real_df.columns]
combined_real_df.to_parquet(f"{cwd}/building_real.parquet")
```
